```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import lagrange
x=np.array([-4.,-2.,1.,3.], dtype=float)
y=np.array([-8.,10.,-8.,20.], dtype=float)
x_0 = -3 # точка, в якій потрібно обчислити значення

def lagrange_interpolation(x, y, x_0):
    n = len(x)
    p = np.zeros(n) # масив для зберігання значень багаточленів L_i
    for i in range(n):
        # обчислюємо багаточлен L_i, який складається з n членів
        # кожен член дорівнює добутку (x-x_j)/(x_i-x_j) для j=0..n, j!=i
        p_i = 1
        for j in range(n):
            if i != j:
                p_i *= (x_0 - x[j])/(x[i] - x[j])
        p[i] = p_i
    return np.dot(y, p) # повертаємо значення багаточлена у точці x_test

# обчислюємо інтерполяційний багаточлен та його значення у точці x_test
f_interp = lagrange_interpolation(x, y, x_0)

print("Значення функції у точці x_0 =", f_interp.round(4))

```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import lagrange
x=np.array([-4.,-2.,1.,3.], dtype=float)
y=np.array([-8.,10.,-8.,20.], dtype=float)
x_3 = 2.5 # точка, в якій потрібно обчислити значення

def lagrange_interpolation(x, y, x_3):
    n = len(x)
    p = np.zeros(n) # масив для зберігання значень багаточленів L_i
    for i in range(n):
        # обчислюємо багаточлен L_i, який складається з n членів
        # кожен член дорівнює добутку (x-x_j)/(x_i-x_j) для j=0..n, j!=i
        p_i = 1
        for j in range(n):
            if i != j:
                p_i *= (x_3 - x[j])/(x[i] - x[j])
        p[i] = p_i
    return np.dot(y, p) # повертаємо значення багаточлена у точці x_test

# обчислюємо інтерполяційний багаточлен та його значення у точці x_test
f_interp = lagrange_interpolation(x, y, x_3)

print("Значення функції у точці x_3 =", f_interp.round(4))

```

```python
xnew=np.linspace(np.min(x),np.max(x),100) #точки, за якими будуємо графік
ynew=[lagrange_interpolation(x,y,i) for i in xnew]

plt.plot(x,y,'o',xnew,ynew) #будуємо графік функції Лагранжа
plt.title('Графік')
plt.grid(True)
plt.show()
```
