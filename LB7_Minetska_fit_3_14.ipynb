```python
import numpy as np

# Функція для обчислення інтерполяційного багаточлена Ньютона
def newton_interpolation(x, y, x_interp):
    n = len(x)
    F = np.zeros((n, n))
    F[:, 0] = y

    for j in range(1, n):
        for i in range(n - j):
            F[i, j] = (F[i + 1, j - 1] - F[i, j - 1]) / (x[i + j] - x[i])

    result = 0
    for j in range(n):
        term = F[0, j]
        for i in range(j):
            term *= (x_interp - x[i])
        result += term

    return result

# Вхідні дані (таблиця значень)
x = np.array([1.415, 1.420, 1.425, 1.430, 1.435, 1.440, 1.445, 1.450, 1.455, 1.460, 1.465])
y = np.array([0.8885, 0.8895,0.8906, 0.8916, 0.8926, 0.8936, 0.8947, 0.8956, 0.8966, 0.8976, 0.8986])

# Точки, в яких ми хочемо обчислити значення функції
x_interp1 = 1.422
x_interp2 = 1.451

# Обчислення значень функції у заданих точках
result1 = newton_interpolation(x, y, x_interp1)
result2 = newton_interpolation(x, y, x_interp2)

print(f'Значення функції у точці {x_interp1} дорівнює {result1}')
print(f'Значення функції у точці {x_interp2} дорівнює {result2}')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Побудова графіку інтерполяційної функції
xx = np.linspace(np.min(x), np.max(x), 100)
yy = np.zeros_like(xx)
for i in range(len(xx)):
    yy[i] = newton_interpolation(x, y, xx[i])

plt.plot(x, y, 'o', label='Дані точки')
plt.plot(xx, yy, label='Багаточлен Ньютона')
plt.title("Графік інтерполяційної функції Ньютона")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```
